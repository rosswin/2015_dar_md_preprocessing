```python
import os
import pandas as pd
```

```python
in_csv = r"/media/ross/ssd/00_data/20191205_HI_2015_coastalimagery_dl/annotations/ALL_final_annotations_20191214.csv"
out_dir = r"/media/ross/ssd/00_data/20191205_HI_2015_coastalimagery_dl/annotations"
```
