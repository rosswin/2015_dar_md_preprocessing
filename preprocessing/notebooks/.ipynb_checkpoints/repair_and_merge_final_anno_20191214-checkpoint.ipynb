```python
import os
import pandas as pd
```

```python
in_dir = r"/media/ross/ssd/00_data/20191205_HI_2015_coastalimagery_dl/annotations"
```

```python
in_csvs = [os.path.join(in_dir, f) for f in os.listdir(in_dir) if f.endswith('.csv')]
```

```python
in_csvs
```

```python
merged_csvs = pd.concat([pd.read_csv(f) for f in in_csvs])
```

```python
merged_csvs
```

```python
#reversing an error in retile_for_dl that did not take into account that geographic software uses a bottom-left origin, and deep learning software uses
#a top left origin.

merged_csvs['new_px_y_max'] = 512 - merged_csvs['px_y_min']
merged_csvs['new_px_y_min'] = 512 - merged_csvs['px_y_max']
```

```python
merged_csvs
```

```python
merged_csvs.to_csv(r"/media/ross/ssd/00_data/20191205_HI_2015_coastalimagery_dl/annotations/ALL_final_annotations_20191214.csv")
```
